```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Parmetros de la seal
A = 1.0  # Amplitud del pulso
tau = 2.0  # Duracin del pulso en segundos
T_total = 10.0  # Duracin total de la seal en segundos
fs = 1000  # Frecuencia de muestreo en Hz (muestras por segundo)
t = np.linspace(-T_total/2, T_total/2, int(T_total * fs), endpoint=False)

# Definir la seal rectangular (Rectangular Pulse)
X_t = A * np.where(np.abs(t) <= tau/2, 1, 0)

# Estimacin de la PSD utilizando el mtodo de Welch
frecuencias, Pxx = welch(X_t, fs, nperseg=256, window='hann', noverlap=128)

# Graficar la seal en el dominio del tiempo
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, X_t)
plt.title("Seal Rectangular en el Tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")

# Graficar la PSD estimada utilizando el mtodo de Welch
plt.subplot(2, 1, 2)
plt.semilogy(frecuencias, Pxx)
plt.title("Densidad Espectral de Potencia (PSD) - Mtodo de Welch")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("PSD [V^2/Hz]")
plt.tight_layout()

# Mostrar los grficos
plt.show()
```
